```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,20,20)
y = x/1.5

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
noise = np.random.normal(0, 1, x.shape)
plt.scatter(x, y+noise)
```

```python
y = y+ noise

learning_rate = 0.001;
max_iter = 10

m = 0.1
b = 1
for i in range(max_iter):
    
    for x_i,y_i in zip(x,y):  #partial derivatives (gradient)
        var_m = (m*x_i + b - y_i)*x_i
        var_b = m*x_i + b - y_i 
        
    m = m - learning_rate*2*var_m
    b = b - learning_rate*2*var_b
```

```python
reg = m*x+b

plt.scatter(x,y,label = 'Original Data')
plt.plot(x,reg,'r--', label = 'Regression Line' )
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.legend()
```
